```julia
using StatisticTools

data=[]
num_data_points = 0
num_of_files = 0
label_column = 0
for file in filter(x->ismatch(r"LW3D24ver[0-9]+\.dat", x), readdir() )
    filedata = readdlm(file);
    label_column = filedata[: ,1]
    data=push!(data, filedata)
    num_data_points = size(filedata,1)
    num_of_files +=1
end

vals=zeros(num_data_points,num_of_files)
errs=zeros(num_data_points,num_of_files)

for i in 1:num_of_files
    vals[:, i]=data[i][:, 2]
    errs[:, i]=data[i][:, 3]
end

finaldata = zeros(num_data_points,3)
for i in 1:num_data_points
    println(vec(vals[i,:]),vec(errs[i,:]))
    w = weightedmean(vec(vals[i,:]),vec(errs[i,:]))
    println(w)
    finaldata[i, 1] = label_column[i]
    finaldata[i, 2] = w[1]
    finaldata[i, 3] = w[2]
end

outfilename = "LW3D24ver.dat"
writedlm(outfilename, finaldata)
```
